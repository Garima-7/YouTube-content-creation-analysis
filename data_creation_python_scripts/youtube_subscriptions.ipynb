```python
import requests
import pandas as pd
```

```python
API_KEY = '<YOUR_API_KEY>'
BASE_URL = "https://www.googleapis.com/youtube/v3"
```

```python
channels_df= pd.read_csv('channels.csv')
```

```python
channel_ids= channels_df['channel_id']
channel_ids
```

```python
endpoint="subscriptions"
subscriptions_url=f"{BASE_URL}/{endpoint}"

all_subscription_details=[]

for channel_id in channel_ids:
    params_subscriptions={
        "key": API_KEY,
        "part":"snippet,contentDetails,subscriberSnippet",
        "channelId":channel_id
    }
    while True:
        response = requests.get(subscriptions_url, params=params_subscriptions)
        data = response.json()

        all_subscription_details.extend(data.get("items", []))

        if "nextPageToken" not in data:
            break

        params_subscriptions["pageToken"] = data["nextPageToken"]
```

```python
subscribers_df = pd.DataFrame({"channel_id":[subscription['snippet']['resourceId']['channelId'] for subscription in all_subscription_details],
                               "subscriber_id":[subscription['snippet']['channelId'] for subscription in all_subscription_details],
                               "subscription_time":[subscription['snippet']['publishedAt'] for subscription in all_subscription_details]})
subscribers_df.to_csv('subscriptions.csv',index=False)
```
