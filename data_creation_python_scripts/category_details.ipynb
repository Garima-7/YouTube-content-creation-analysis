```python
import pandas as pd
import requests
```

```python
videos_df = pd.read_csv('../videos.csv')
```

```python
len(videos_df)
category_ids = set(videos_df['category_id'])
```

```python
list(category_ids)
```

```python
API_KEY = '<YOUR_API_KEY>'
BASE_URL = "https://www.googleapis.com/youtube/v3"
```

```python
endpoint= "videoCategories"
category_url = f"{BASE_URL}/{endpoint}"
params= {
    "key":API_KEY,
    "part":'snippet',
    "id": category_ids,
}
```

```python
response = requests.get(category_url,params=params)
```

```python
channel_category_details=[]

for detail in response.json()['items']:
    channel_category_details.append(detail)
```

```python
category_df = pd.DataFrame({
    "category_id" : [category['id'] for category in channel_category_details],
    "category_name" : [category['snippet']['title'] for category in channel_category_details]
})
```

```python
category_df.to_csv('../category.csv')
```
